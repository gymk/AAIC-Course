```python
# https://ndres.me/post/matplotlib-animated-gifs-easily/

import random
%matplotlib inline

frames = []
frame_no = 0

def createFrame(frame_no, y_max):

    x = [random.randint(1,50) for x in range(50)]
    y = [random.randint(1,50) for x in range(50)]

    fig, ax = plt.subplots(figsize=(10,5))
    #ax.plot(x, y)
    ax.scatter(x,y)
    ax.set(xlabel='x', ylabel='y', title='{}'.format(frame_no))
    
    ax.set_ylim(0, y_max)
    
    fig.canvas.draw()
    image = np.frombuffer(fig.canvas.tostring_rgb(), dtype='uint8')
    image  = image.reshape(fig.canvas.get_width_height()[::-1] + (3,))

    frames.append(image)
    
# Create Frames
for i in range(60):
    createFrame(i+1, 50)

# Create Animation
kwargs_write = {'fps':1.0, 'quantizer':'nq'}
imageio.mimsave('./powers-2.gif', frames, fps=1)
    
```
