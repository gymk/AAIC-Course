# t-SNE on MNIST Handwritten Digits Dataset

```python
# import required modules
import numpy as np # for matrix multiplication
import matplotlib.pyplot as plt # for plotting
import seaborn as sns # for plotting
import pandas as pd # for data handling
from sklearn.preprocessing import StandardScaler # for column standardization
from sklearn.manifold import TSNE # for t-SNE
```

```python
# Read dataset
d0 = pd.read_csv('./../../AAIC-Course/datasets/mnist-digits-dataset/train.csv')
d0.head()
```

```python
# split the dataset into features and labels
d_labels = d0['label']
d_data = d0.drop('label',axis=1)
print(d_labels.shape,d_data.shape)
```

```python
# standardize the data
standardized_data = StandardScaler().fit_transform(d_data.astype(np.float64))
standardized_data.shape
```

## t-SNE over 1000 data points

```python
data_1000 = standardized_data[:1000]
print(data_1000.shape)
```

### Perplexity = default (30), Iterations = default (1000)

```python
# Compute t-SNE (784-D to 2-D) for visualization
model = TSNE(n_components=2,random_state=0)

tsne_data = model.fit_transform(data_1000)
print(tsne_data.shape)

tsne_data = np.vstack((tsne_data.T,d_labels[:1000])).T
print(tsne_data.shape)
tsne_data[:4]

tsne_df = pd.DataFrame(tsne_data,columns=['Dim_1','Dim_2','label'])
tsne_df.head()

sns.FacetGrid(tsne_df,hue='label',height=7).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
```

### Perplexity = default (30), Iterations = 5000

```python
# Compute t-SNE (784-D to 2-D) for visualization
data_1000 = standardized_data[:1000]
print(data_1000.shape)

model = TSNE(n_components=2,random_state=0,n_iter=5000)

tsne_data = model.fit_transform(data_1000)
print(tsne_data.shape)

tsne_data = np.vstack((tsne_data.T,d_labels[:1000])).T
print(tsne_data.shape)
tsne_data[:4]

tsne_df = pd.DataFrame(tsne_data,columns=['Dim_1','Dim_2','label'])
tsne_df.head()

sns.FacetGrid(tsne_df,hue='label',height=7).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
```

### Perplexity = default (30), Iterations = 10000

```python
# Compute t-SNE (784-D to 2-D) for visualization
data_1000 = standardized_data[:1000]
print(data_1000.shape)

model = TSNE(n_components=2,random_state=0,n_iter=10000)

tsne_data = model.fit_transform(data_1000)
print(tsne_data.shape)

tsne_data = np.vstack((tsne_data.T,d_labels[:1000])).T
print(tsne_data.shape)
tsne_data[:4]

tsne_df = pd.DataFrame(tsne_data,columns=['Dim_1','Dim_2','label'])
tsne_df.head()

sns.FacetGrid(tsne_df,hue='label',height=7).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
```
